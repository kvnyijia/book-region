```python
import pandas as pd
import math
```

```python
path = '../Book-Crossing-User-review-ratings/Book-reviews/BX-Book-Ratings.csv'
df1 = pd.read_csv(path, sep=';', encoding='latin-1', low_memory=False)
path = '../Book-Crossing-User-review-ratings/Book-reviews/BX-Users.csv'
df2 = pd.read_csv(path, sep=';', encoding='latin-1', low_memory=False)
path = '../Book-Crossing-User-review-ratings/Book-reviews/BX_Books.csv'
df3 = pd.read_csv(path, sep=';', encoding='latin-1', low_memory=False)
path = '../Book-Crossing-User-review-ratings/Books-Data-with-Category-Language-and-Summary/Preprocessed_data.csv'
df4 = pd.read_csv(path)
```

```python
dfg = df4.groupby(['country', 'isbn', 'img_l']).agg({'user_id': 'count', 'rating': 'mean'})
dfg['log_user_id_cnt'] = dfg['user_id'].apply(lambda x: math.log(x, 2))

user_cnt_max = dfg['log_user_id_cnt'].max()
rating_max = dfg['rating'].max()
dfg['weight'] = dfg[['log_user_id_cnt', 'rating']].apply(lambda x: x[0] / user_cnt_max * 70 + x[1] / rating_max * 30, axis=1)
dfg
```

```python
dfsort = dfg.sort_values(by='weight', ascending=False)
```

```python
dfsort.groupby(['country', 'isbn', 'img_l']).agg({'weight': 'sum'}).reset_index()
```

```python
dfg = df4.groupby(['country', 'isbn', 'img_l', 'book_title']).agg({'user_id': 'count', 'rating': 'mean'})
dfg['log_user_id_cnt'] = dfg['user_id'].apply(lambda x: math.log(x, 2))

user_cnt_max = dfg['log_user_id_cnt'].max()
rating_max = dfg['rating'].max()
dfg['weight'] = dfg[['log_user_id_cnt', 'rating']].apply(lambda x: x[0] / user_cnt_max * 70 + x[1] / rating_max * 30, axis=1)
# dfg['weight'] = dfg['weight'].apply(lambda x: int(x))
```

```python
g = dfg['weight'].groupby('country', group_keys=False)
```

```python
res = g.apply(lambda x: x.sort_values(ascending=False))
```

```python
res = res.reset_index()
```
